```python
import numpy as np

# Dataset
X = np.array([
    [1, 1.5],
    [2, 2.0],
    [3, 2.5],
    [4, 3.0]
])

y = np.array([0, 0, 1, 1])

# Initialize weights and bias
weights = np.random.rand(2)
bias = 0.1
lr = 0.1  # learning rate

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Training loop
for epoch in range(1000):
    for i in range(len(X)):
        z = np.dot(X[i], weights) + bias
        y_pred = sigmoid(z)

        error = y[i] - y_pred

        weights += lr * error * X[i]
        bias += lr * error

# Prediction
test_input = np.array([3, 2.8])
output = sigmoid(np.dot(test_input, weights) + bias)

print("Predicted Health Status:", int(output > 0.5))
```

```python
import numpy as np

# ---------------- DATASET ----------------
# Input features: [assignments submitted, attendance percentage]
X = np.array([
    [3, 60],
    [5, 70],
    [7, 85],
    [9, 90]
])

# Output labels: Pass (1) or Fail (0)
y = np.array([0, 0, 1, 1])

# ---------------- INITIALIZATION ----------------
# Random weights for two input features
weights = np.random.rand(2)

# Bias value
bias = 0.1

# Learning rate
lr = 0.01

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# ---------------- TRAINING ----------------
# Train the ANN using simple gradient update
for epoch in range(1000):
    for i in range(len(X)):
        z = np.dot(X[i], weights) + bias
        y_pred = sigmoid(z)

        error = y[i] - y_pred

        # Update weights and bias
        weights += lr * error * X[i]
        bias += lr * error

# ---------------- PREDICTION ----------------
# New student data
test_input = np.array([8, 88])

output = sigmoid(np.dot(test_input, weights) + bias)

print("Predicted Result (Pass=1, Fail=0):", int(output > 0.5))
```

```python
import numpy as np

# ---------------- DATASET ----------------
# Input features: [monthly income, credit score]
X = np.array([
    [30, 550],
    [40, 600],
    [60, 700],
    [80, 750]
])

# Output labels
y = np.array([0, 0, 1, 1])

# ---------------- INITIALIZATION ----------------
weights = np.random.rand(2)
bias = 0.1
lr = 0.0001  # small learning rate (large values in data)

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# ---------------- TRAINING ----------------
for epoch in range(1500):
    for i in range(len(X)):
        z = np.dot(X[i], weights) + bias
        y_pred = sigmoid(z)

        error = y[i] - y_pred

        weights += lr * error * X[i]
        bias += lr * error

# ---------------- PREDICTION ----------------
# New applicant
test_input = np.array([70, 720])

output = sigmoid(np.dot(test_input, weights) + bias)

print("Loan Eligibility (1=Yes, 0=No):", int(output > 0.5))
```
