```python
import heapq
graph = {
    0: [1, 2],
    1: [3, 4, 5],
    2: [5, 6],
    3: [],
    4: [],
    5: [],
    6: []
}
def ucs(graph, start, goal):
    pq = [(0, start, [])]
    visited = set()
    while pq:
        cost, node, path = heapq.heappop(pq)
        if node in visited:
            continue
        visited.add(node)
        path = path + [node]
        if node == goal:
            print("Path:", path, "Cost:", cost)
            return
        for child in graph[node]:
            heapq.heappush(pq, (cost + 1, child, path))
ucs(graph, 0, 5)
```

```python
from collections import deque
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['I', 'J'],
    'D': ['H'],
    'E': ['H'],
    'F': ['C', 'I'],
    'H': ['J'],
    'I': ['J', 'G'],
    'J': ['G'],
    'G': []
}
def bfs(start, goal):
    visited = set()
    queue = deque([(start, [start])])
    while queue:
        node, path = queue.popleft()
        if node == goal:
            return path
        if node not in visited:
            visited.add(node)
            for neighbour in graph[node]:
                queue.append((neighbour, path + [neighbour]))

    return None
print("BFS:", bfs('A', 'G'))
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['I', 'J'],
    'D': ['H'],
    'E': ['H'],
    'F': ['C', 'I'],
    'H': ['J'],
    'I': ['J', 'G'],
    'J': ['G'],
    'G': []
}
def dfs(start, goal):
    visited = set()
    stack = [(start, [start])]
    while stack:
        node, path = stack.pop()
        if node == goal:
            return path
        if node not in visited:
            visited.add(node)
            for neighbour in reversed(graph[node]):
                stack.append((neighbour, path + [neighbour]))
    return None
print("DFS:", dfs('A', 'G'))
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['I', 'J'],
    'D': ['H'],
    'E': ['H'],
    'F': ['C', 'I'],
    'H': ['J'],
    'I': ['J', 'G'],
    'J': ['G'],
    'G': []
}
def dls(node, goal, limit, path):
    if node == goal:
        return path
    if limit == 0:
        return None
    for neighbour in graph[node]:
        result = dls(neighbour, goal, limit - 1, path + [neighbour])
        if result is not None:
            return result
    return None
print("DLS:", dls('A', 'G', 4, ['A']))
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['I', 'J'],
    'D': ['H'],
    'E': ['H'],
    'F': ['C', 'I'],
    'H': ['J'],
    'I': ['J', 'G'],
    'J': ['G'],
    'G': []
}
def dls(node, goal, limit, path):
    if node == goal:
        return path
    if limit == 0:
        return None
    for neighbour in graph[node]:
        result = dls(neighbour, goal, limit - 1, path + [neighbour])
        if result:
            return result
    return None
def iddfs(start, goal, max_depth=10):
    for depth in range(max_depth):
        result = dls(start, goal, depth, [start])
        if result:
            return result
    return None
print("IDDFS:", iddfs('A', 'G'))
```

```python
import heapq
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['I', 'J'],
    'D': ['H'],
    'E': ['H'],
    'F': ['C', 'I'],
    'H': ['J'],
    'I': ['J', 'G'],
    'J': ['G'],
    'G': []
}
def ucs(start, goal):
    pq = []
    heapq.heappush(pq, (0, start, [start]))
    visited = {}
    while pq:
        cost, node, path = heapq.heappop(pq)

        if node == goal:
            return path, cost

        if node not in visited or cost < visited[node]:
            visited[node] = cost
            for neighbour in graph[node]:
                heapq.heappush(pq, (cost + 1, neighbour, path + [neighbour]))
    return None
print("UCS:", ucs('A', 'G'))
```
